```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread("./tmp_4.jpg", cv2.IMREAD_COLOR)#[300:550, 750:1000]
plt.imshow(img[:,:,::-1])
```

```python
img_hls = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
npc_mask = cv2.inRange(img_hls, (0/2, 0.52*255, 0.01*255), (360/2, 0.59*255, 0.06*255))
plt.imshow(npc_mask, cmap="gray")
```

```python
img_circle = img.copy()
cnts = cv2.findContours(npc_mask[500:], cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
print(len(cnts))  # number of contours
# cnts = cnts[0] if len(cnts) == 2 else cnts[1]
cnts = cnts[0]# if len(cnts) == 2 else cnts[1]
print(len(cnts))  # number of contours
for c in cnts:
    # peri = cv2.arcLength(c, True)
    # approx = cv2.approxPolyDP(c, 0.04 * peri, True)
    area = cv2.contourArea(c)
    # print(len(approx), area)
    # if len(approx) > 5 and area > 100 and area < 500:
    if area > 2000:
        print(area)
        ((x, y), r) = cv2.minEnclosingCircle(c)
        cv2.circle(img_circle, (int(x), int(y)), int(r), (0, 255, 0), 2)

plt.imshow(img_circle[:,:,::-1], cmap="gray")
```
